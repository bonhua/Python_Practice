## **ANALYZE THE DATA**

### 9. Create a crosstab, with category_name in rows and country in columns.

```python
crosstab = pd.crosstab(northwind["category_name"], northwind["country"])
crosstab
```

### 10. Create a pivot table with average unit_price by category.

```python
avg_unitprice = northwind.pivot_table(index="category_name", values="unit_price")
avg_unitprice
```

### 11. Create a pivot table with average reorder_level  by category (rows) and country (columns).

```python
avg_reorder_level = northwind.pivot_table(index="category_name", values="reorder_level", columns="country")
avg_reorder_level
```

### 12. Create a groupby table with sum of discontinued products by company name.

```python
discontinued = northwind.groupby("company_name")["discontinued"].sum()
discontinued
```

### 13. Create a groupby table with maximum unit_price by country and category (both in rows).

```python
maxprice = northwind.groupby(["country","category_name"])["unit_price"].max()
maxprice
```
